```python
def isConvex(p1,p2,p3): #return true if convex or straight 
    a = np.array(p1)
    b = np.array(p2)
    c = np.array(p3) 
#a = np.array([1, 4])
#b = np.array([8, 4])
#c = np.array([7, 3]) 
    ab=a-b
    cb=c-b
#print(ab)
#print(cb)
    cross = np.cross(cb,ab) #negative degree -> convex 凸                                                         #https://stackoverflow.com/questions/20252845/best-algorithm-for-detecting-interior-and-exterior-angles-of-an-arbitrary-shape
    dot = np.inner(cb,ab)
    radtan = np.arctan2(cross,dot)
    degtan = np.rad2deg(radtan)
    #print(degtan)
    return degtan<=0
```

```python
def newcoord(coords, dist): #coords: (A,B) -- from A to B. Extend after B.      #啊啊啊python真的太难受了！！x1是什么，是什么type啊！！
    (x1,y1),(x2,y2) = coords 
    dx = x2 - x1
    dy = y2 - y1
    linelen = math.hypot(dx, dy)

    x3 = x2 + dx/linelen * dist
    y3 = y2 + dy/linelen * dist    
    return x3, y3

n = newcoord((clean1[1],clean1[2]),0.1)  
p2 = Point(n)                                                 # can't believe I wrote the wrong variable name and debugged for half an hour...
p2.within(cleanpolygon1) #parallel does not count as inside polygon 
p2.within(border) #if point is within inner polygon, it is within border 
```

```python
def findreflex(before):
    r1=[]
    for i in range(len(before)):
        if isConvex(before[i-1],before[i],before[((i+1)%(len(before)))]):
            r1.add(before[i])
    return r1
```
